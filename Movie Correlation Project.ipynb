```python
# import the data packages that we are going to use - for this task. the movie dataset is taken from excel
# and the kaggle website we are trying to find out the correlation between budget and gross and
# we are going to out if a films budget and their  gross is somewhat related - films are dated from 1980
# all the way to 2020

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in the data

df = pd.read_csv(r'/Users/tenikuti/Downloads/movies2.csv')
```

```python


# Looking at the top 15 compaies by gross revenue
# Warmner bros has the highest gross revenue between 1980 - 2012 out of all the companies selected

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted

```

```python
#  you can see there is somewhat of a correlatoion betwween budget vs gross earnings

plt.scatter(x=df['budget'], y=df['gross'], alpha=0.5)
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# To further  examine what we can do is we do a regression plot using a seaborn package to do a line of best fit


sns.regplot(x="gross", y="budget", data=df)


```
